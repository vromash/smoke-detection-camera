```
# Utils
def in_darknet(func):
  def wrapper_in_darknet(*args, **kwargs):
    import os, sys
    if os.getcwd() != '/content/darknet':
      os.chdir('/content/darknet')
    try:
      return func(*args, **kwargs)
    except:
      e = sys.exc_info()[0]
      print('Error has occured', e)
    finally:
      os.chdir('/content')
  return wrapper_in_darknet

def download_video(path):
  from google.colab import files
  files.download(path)

def prepare_stats(path):
  video_data = get_video_data(path)
  return {
      'frame_number': 0,
      'fps': video_data[1]
  }

def get_video_data(path):
  import cv2
  cap = cv2.VideoCapture(path)
  length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
  fps = int(cap.get(cv2.CAP_PROP_FPS))
  return (length, fps)
```

```
# Main functions
@in_darknet
def test_image(image_path, print_output = False):
  import shlex, subprocess, os

  command = f'./darknet detector test \
    build/darknet/x64/data/obj.data \
    cfg/yolov4-tiny-obj.cfg \
    build/darknet/x64/backup/yolov4-tiny-obj_last.weights {image_path} -i 0 -thresh 0.20'
  
  process = subprocess.Popen(shlex.split(command), stdout=subprocess.PIPE)
  process_alive = True
  while process_alive:
    output = process.stdout.readline()
    if print_output:
      print(output)
    if output == b'' and process.poll() is not None:
      process_alive = False

  im_show('predictions.jpg')
  os.remove('predictions.jpg')


@in_darknet
def test_video(video_path, print_output = False, gather_stats = True, download_on_success = False):
  import shlex, subprocess, os

  output_video = 'result-video.mp4'
  command = f'./darknet detector demo \
    build/darknet/x64/data/obj.data \
    cfg/yolov4-tiny-obj.cfg \
    build/darknet/x64/backup/yolov4-tiny-obj_last.weights \
    -thresh 0.25 -dont_show {video_path} -out_filename {output_video}'

  if gather_stats:
    stats = prepare_stats(video_path)

  process = subprocess.Popen(shlex.split(command), stdout=subprocess.PIPE)
  process_alive = True

  while process_alive:
    output = process.stdout.readline()
    if print_output:
      print(output)

    if gather_stats:
      parse_line(output, ['smoke', 'cigarete'], stats)

    if output == b'' and process.poll() is not None:
      process_alive = False

  if gather_stats:
    print_stats(stats)
  
  if download_on_success:
    download_video(output_video)

  # os.remove(output_video)
  return stats
```

```
# Output parser
def parse_line(output, targets, stats):
  import re
  str_output = str(output, 'utf-8')
  success_limit = 70

  item = {
      'name': 'frame'
  }

  if 'cvWriteFrame' in str_output:
    update_statistics(stats, item)
    return

  for class_name in targets:
    if class_name in str_output:
      percentage = re.findall('[0-9][0-9]', str_output)
      if int(percentage[0]) < success_limit:
        return
      
      item['name'] = class_name
      item['percentage'] = int(percentage[0])
      update_statistics(stats, item)

def update_statistics(stats, item):
  if item['name'] == 'frame':
    stats['frame_number'] += 1
  else:
    frame = stats['frame_number']
    second = round(stats['frame_number'] / stats['fps'])
    data = item['name'] + ': ' + str(item['percentage']) + ' %'
    if second in stats:
      stats[second][frame] = data
    else:
      stats[second] = {}
      stats[second][frame] = data
```

```
# Vizualization
def im_show(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()

def print_stats(stats):
  import datetime
  for item in stats:
    if item == 'frame_number':
      print('Frame number: ' + str(stats[item]))
    elif item == 'fps':
      print('FPS: ' + str(stats[item]))
      print('-----------------------------------')
    else:
      print(str(datetime.timedelta(seconds=item)))
      print('\tFrame\tObject')
      for frame in stats[item]:
        print('\t' + str(frame) + '\t' + str(stats[item][frame]))
```

```
# Get requested frame
@in_darknet
def get_frame(video_path, frame_id):
  import cv2, os

  vidcap = cv2.VideoCapture(video_path)
  fps = int(vidcap.get(cv2.CAP_PROP_FPS))
  rate = 1 / fps * 1000

  vidcap.set(cv2.CAP_PROP_POS_MSEC, frame_id * rate)

  hasFrames, image = vidcap.read()
  if hasFrames:
    cv2.imwrite('requested_frame.jpg', image)
    im_show('requested_frame.jpg')
    os.remove('requested_frame.jpg')
  else:
    print('Frame doesn\'t exist')  
```
